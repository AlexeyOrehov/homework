# Итоговый проект.

Анализ личности клиента — это подробный анализ клиентов компании. Он помогает компании лучше понять своих клиентов и облегчает изменение продукции в соответствии со специфическими потребностями, поведением различных типов клиентов.

Анализ личности клиента помогает компании модифицировать свой продукт в соответствии с целевыми потребителями из различных сегментов потребителей. Например, вместо того, чтобы тратить деньги на маркетинг нового продукта для каждого клиента в базе данных компании, компания может проанализировать, какой сегмент клиентов с наибольшей вероятностью купит продукт, и затем продвигать продукт только для этого сегмента.

Описание столбцов:

|--|--|

Для правильного анализа трат на вино, предлагаю сегментировать клиентов на основе их трат на вино, потому что медианное значение трат на вино - ~200, а среднее - 304, что говорит нам, что есть люди, и их большинство, которые тратят на вино мало, о чем нам указывает медианное значение, но есть и клиенты, которые тратят на вино много - эта группа людей меньше, но они вносят значительный вклад в общую выручку

```python
def segment_by_mnt(total):
    if total <= 69:
        return 'Low spenders'
    elif total <= 396.5:
        return 'Medium-low spenders'
    elif total <= 1044.25:
        return 'Medium spenders'
    else:
        return 'High spenders'
    
df_modify['mnt_segment'] = df_modify['mnt'].apply(segment_by_mnt)
```

```python
df_modify['mnt_segment'].value_counts()
```

Сделал сегментацию по квартилиям от общей суммы трат клиентов

```python
df_modify.groupby('mnt_segment')['mntwines'].agg(['mean', 'median'])
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_modify, x='mnt_segment', y='mntwines')
plt.title('Распределение трат на вино по сегментам общих трат')
plt.xlabel('Сегмент клиентов по общим тратам')
plt.ylabel('Сумма на вино')
plt.show()
```

__Средние траты сегментов на вино:__
- Мало-тратящие - 11, более активно тратящие - 70, средне-тратящие - 356, много-тратящие - 737, брал везде медианное значение, так как во всех группах, за исключением много-тратящих, есть выбросы
